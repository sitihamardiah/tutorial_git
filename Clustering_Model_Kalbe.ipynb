## **K-Means Clustering** ##

**Segmentasi berdasarkan Umur dan Harga Pembelian**

**Menentukan K (banyaknya kelompok) yang optimal**

```python
X1=df[['umur', 'harga_pembelian']]
wcss=[]
for n in range(1, 11):
    model1 = KMeans(n_clusters=n, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=100)
    model1.fit(X1)
    wcss.append(model1.inertia_)
print(wcss)
```

### **Elbow Method** ###

```python
plt.figure(figsize=(8,3))
plt.plot(list(range(1, 11)), wcss, color='blue', marker='o', linewidth=2, markersize=12, markerfacecolor='m', markeredgecolor='m')
plt.title('WCSS VS Banyaknya Cluster', fontsize=18)
plt.xlabel('Jumlah Cluster', fontsize=14)
plt.ylabel('WCSS', fontsize=14)
plt.show()
```

**Observations:** \

Berdasarkan Elbow Method diatas, Cluster yang paling tepat adalah 3

### **Membuat Model Clustering dengan K yang optimal** ###

```python
model1 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=100)
model1.fit(X1)
labels1 = model1.labels_
centroids1 = model1.cluster_centers_
```

### **Customer Segmentation berdasarkan Umur dan Harga Pembelian** ###

```python
plt.figure(figsize=(11,6))
sns.set_style('white')
plt.scatter(x=df['umur'], y=df['harga_pembelian'], c=labels1, cmap='winter')
plt.scatter(x=centroids1[: , 0], y=centroids1[: , 1], s=300, c='red')
plt.xlabel('umur', fontsize=14)
plt.ylabel('harga pembelian', fontsize=14)
plt.title('Customer Segmentation berdasarkan Umur dan Harga Pembelian')
plt.show()
```
